```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
df=pd.read_csv("C:\\Users\\Swetanshu\\Desktop\\Lung_cancer.csv")
print(df)
```

"Visualize the count of patients by treatment type to analyze the most commonly used treatments for lung cancer based on BMI."

```python
treat_count=df['treatment_type'].value_counts()
plt.figure(figsize=(10,6))
sns.barplot(x=treat_count.index,y=treat_count.values,palette='viridis')
plt.title("Treatment and understanding of the paitent")
plt.xlabel("treatment_type")
plt.ylabel("bmi")
plt.tight_layout()
plt.show()
```

Create a bar graph to compare survival rates across different cancer stages.

```python
survival_counts=df.groupby('cancer_stage')['survived'].value_counts()
survival_counts.plot(kind="bar",figsize=(10,6),color="orange")
plt.title("Cancer stage and survive")
plt.xlabel("cancer_stage")
plt.ylabel("survived")
plt.tight_layout()
plt.show()
```
